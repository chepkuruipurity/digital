```python
# load the libraries urllib and json
from urllib.request import urlopen
import json
```

```python
#url
url = 'https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json'
```

```python
# hold the response of url
response= urlopen(url)
```

```python
#load the json data
data_json =json.load(response)
```

```python
#final data structure
final ={}
```

```python
# find the functional water points
def count(arr):
    n=0
    k=0
    for data in arr:
        if data['water_point_condition'] == 'functioning':
            n  += 1 
    final['number_functional']= n    
    return n
    
    
```

```python
#redering the count function
count(data_json)
```

```python
#finding how many communities have water points/ number of water points in communities
def freq(arr):
    
    freq={}
    for data in  arr:
        if (data['communities_villages'] in freq):
            freq[data['communities_villages']] += 1
        else:
            freq[data['communities_villages']] = 1
            
    for key, value in freq.items():
        print (key,value)
    final['number_water_points'] = freq
 
```

```python
freq(data_json)
```

```python
# finding community rank based on missing functional water points
def func(arr):
  
    rank = {}
    new={}
    for data in arr:
        if data['water_functioning'] == 'no'and data['communities_villages'] in rank:
                    rank[data['communities_villages']] += 1
        elif data['water_functioning']== 'no' and data['communities_villages'] not in rank :
            rank[data['communities_villages']] = 1
        elif data['water_functioning']== 'yes' and data['communities_villages'] not in rank :
            rank[data['communities_villages']] = 0
      
               
    for key , value in rank.items():
        newrank= '{:.2f}'.format(value/len(arr)*100) + ' %'
        print(key,newrank)
        new[key]= newrank
    final['community_ranking']= new
    
        
   
```
